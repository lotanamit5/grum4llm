# Reproducing Original GRUMs Figures
This notebook recreates Figures 2, 3, 4, 5, and 6 natively analyzing the active evaluation datasets evaluated through `asymptotic_repro` and `elicitation_repro` Slurm configurations.

```
import sys
from pathlib import Path
ROOT = Path('.').resolve().parents[0]
if str(ROOT / 'src') not in sys.path:
    sys.path.insert(0, str(ROOT / 'src'))

import matplotlib.pyplot as plt
from grums.utils.plot_utils import load_metrics_for_datasets
from grums.utils.plots import plot_asymptotic, plot_elicitation

# Define canonical roots corresponding to orchestration execution folders
ASYM_DIR = ROOT / "results/repro/asymptotic_repro"
ELIC_DIR = ROOT / "results/repro/elicitation_repro"
```

## Figure 2: Social Choice Correlation (Asymptotic)

```
# Dataset0: social-corr metric over 500 steps. Evaluates via native step boxes mapping all seeds efficiently.
df_asym = load_metrics_for_datasets(ASYM_DIR, datasets=["ds0"], pivot_metrics=True)
if not df_asym.empty:
    fig, ax = plt.subplots(figsize=(10, 6))
    plot_asymptotic(df_asym, metrics_columns=["social_tau"], ax=ax)
    ax.set_title("Figure 2: Social Choice Consistency")
    plt.show()
else:
    print("Data not found for Figure 2. Run the asymptotic_orchestration.yml pipeline first.")
```

## Figure 6: Personalized Choice Correlation (Asymptotic)

```
# Dataset0: tracks 'mean_person_tau' natively globally plotting boxes identically alongside 'raw_person_tau'.
if not df_asym.empty:
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(18, 6))
    plot_asymptotic(df_asym, metrics_columns=["mean_person_tau"], ax=ax1)
    ax1.set_title("Figure 6a: Personalized Pred. Correlation")
    
    plot_asymptotic(df_asym, metrics_columns=["raw_person_tau"], ax=ax2)
    ax2.set_title("Figure 6b: Personalized Raw Correlation")
    plt.show()
```

## Figure 3: Social Choice Correlation (Elicitation)

```
# Datasets ds1, ds2. Evaluates exactly line boundaries over 90 iterations tracking 'value', 'criteria', 'seed', 'step'.
df_elic_ds_soc = load_metrics_for_datasets(ELIC_DIR, datasets=["ds1", "ds2"], metric_filter="social_tau")

if not df_elic_ds_soc.empty:
    fig, ax = plt.subplots(figsize=(8, 6))
    plot_elicitation(df_elic_ds_soc, window_size=10, ax=ax)
    ax.set_title("Figure 3: Social Choice Elicitation")
    plt.show()
else:
    print("Data not found for Figure 3. Run the elicitation_orchestration.yml pipeline first.")
```

## Figure 4: Personalized Choice Correlation (Elicitation)

```
# Datasets ds1, ds2. Identical configuration plotting mean personalized tau cleanly mapping interval measurements
df_elic_ds_pers = load_metrics_for_datasets(ELIC_DIR, datasets=["ds1", "ds2"], metric_filter="mean_person_tau")

if not df_elic_ds_pers.empty:
    fig, ax = plt.subplots(figsize=(8, 6))
    plot_elicitation(df_elic_ds_pers, window_size=10, ax=ax)
    ax.set_title("Figure 4: Personalized Choice Elicitation")
    plt.show()
```

## Figure 5: Sushi Application

```
# Data: sushi exclusively. Analyzes social and mean personalized criteria natively alongside identical interval evaluation benchmarks.
df_sushi_soc = load_metrics_for_datasets(ELIC_DIR, datasets=["sushi"], metric_filter="social_tau")
df_sushi_pers = load_metrics_for_datasets(ELIC_DIR, datasets=["sushi"], metric_filter="mean_person_tau")

if not df_sushi_soc.empty and not df_sushi_pers.empty:
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 6))
    plot_elicitation(df_sushi_soc, window_size=10, ax=ax1)
    ax1.set_title("Figure 5a: Sushi Social Choice")
    
    plot_elicitation(df_sushi_pers, window_size=10, ax=ax2)
    ax2.set_title("Figure 5b: Sushi Personalized Choice")
    plt.show()
```
