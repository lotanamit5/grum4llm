# Reproducing Original GRUMs Figures

```python
import sys
from pathlib import Path
import pandas as pd
import matplotlib.pyplot as plt

ROOT = Path('.').resolve().parents[0]
if str(ROOT / 'src') not in sys.path:
    sys.path.insert(0, str(ROOT / 'src'))

from grums.utils.plot_utils import load_metrics_dataframe
from grums.utils.plots import plot_asymptotic, plot_elicitation

ASYM_DIR = ROOT / "artifacts/repro/asymptotic"
ELIC_DIR = ROOT / "artifacts/repro/elicitation"

# Load dataframes
asym_df = load_metrics_dataframe(ASYM_DIR)
elic_df = load_metrics_dataframe(ELIC_DIR)
```

## Figure 2: Asymptotic Behavior, Synthetic Data, Social Choice

```python
if not asym_df.empty:
    fig2_df = asym_df[asym_df["criteria"] == "social"]
    fig, ax = plt.subplots(figsize=(10, 6))
    plot_asymptotic(fig2_df, metrics_columns=["social_tau"], ax=ax)
    fig.suptitle("Asymptotic Behavior, Synthetic Data, Social Choice")
    plt.tight_layout()
    plt.show()
```

## Figure 3: Synthetic Data, Social Choice

```python
if not elic_df.empty:
    fig3_df = elic_df[(elic_df["dataset"].isin(["ds1", "ds2"])) & (elic_df["criteria"] != "personalized")]
    
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 6))
    fig.suptitle("Synthetic Data, Social Choice")
    
    plot_elicitation(fig3_df[fig3_df["dataset"] == "ds1"], metric_col="social_tau", window_size=10, ax=ax1)
    ax1.set_title("Dataset 1 (ds1)")
    
    plot_elicitation(fig3_df[fig3_df["dataset"] == "ds2"], metric_col="social_tau", window_size=10, ax=ax2)
    ax2.set_title("Dataset 2 (ds2)")
    
    plt.tight_layout()
    plt.show()
```

## Figure 4: Synthetic Data, Personalized Choice

```python
if not elic_df.empty:
    fig4_df = elic_df[(elic_df["dataset"].isin(["ds1", "ds2"])) & (elic_df["criteria"] != "social")]
    
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 6))
    fig.suptitle("Synthetic Data, Personalized Choice")
    
    plot_elicitation(fig4_df[fig4_df["dataset"] == "ds1"], metric_col="person_tau", window_size=10, ax=ax1)
    ax1.set_title("Dataset 1 (ds1)")
    
    plot_elicitation(fig4_df[fig4_df["dataset"] == "ds2"], metric_col="person_tau", window_size=10, ax=ax2)
    ax2.set_title("Dataset 2 (ds2)")
    
    plt.tight_layout()
    plt.show()
```

## Figure 5: Sushi Dataset

```python
if not elic_df.empty:
    fig5_df = elic_df[elic_df["dataset"] == "sushi"]
    
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 6))
    fig.suptitle("Sushi Dataset")
    
    # Social Choice: filter out personalized
    plot_elicitation(fig5_df[fig5_df["criteria"] != "personalized"], metric_col="social_tau", window_size=10, ax=ax1)
    ax1.set_title("Social Choice")
    
    # Personalized Choice: filter out social
    plot_elicitation(fig5_df[fig5_df["criteria"] != "social"], metric_col="person_tau", window_size=10, ax=ax2)
    ax2.set_title("Personalized Choice")
    
    plt.tight_layout()
    plt.show()
```

## Figure 6: Asymptotic Behavior, Synthetic Data, Personalized Choice

```python
if not asym_df.empty:
    fig, ax = plt.subplots(figsize=(12, 6))
    plot_asymptotic(asym_df, metrics_columns=["person_tau", "raw_tau"], ax=ax)
    fig.suptitle("Asymptotic Behavior, Synthetic Data, Personalized Choice")
    plt.tight_layout()
    plt.show()
```
